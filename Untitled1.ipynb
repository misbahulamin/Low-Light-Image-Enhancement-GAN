```python
import os
import numpy as np
from PIL import Image

def load_images_from_folder(folder, target_size=(256, 256)):
    images = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = Image.open(img_path).resize(target_size)
        img = np.array(img) / 255.0  # Normalize to [0, 1]
        images.append(img)
    return np.array(images)

# Set the dataset folder path
dataset_folder = './lol_dataset/'

# Paths to training and validation data
train_low_folder = os.path.join(dataset_folder, 'eval15', 'low')
train_high_folder = os.path.join(dataset_folder, 'eval15', 'high')
val_low_folder = os.path.join(dataset_folder, 'our485', 'low')
val_high_folder = os.path.join(dataset_folder, 'our485', 'high')

# Load images
train_low = load_images_from_folder(train_low_folder)
train_high = load_images_from_folder(train_high_folder)
val_low = load_images_from_folder(val_low_folder)
val_high = load_images_from_folder(val_high_folder)

print(f"Training data shapes: {train_low.shape}, {train_high.shape}")
print(f"Validation data shapes: {val_low.shape}, {val_high.shape}")
```

```python
print("Train Low Images:", os.listdir(train_low_folder))
print("Train High Images:", os.listdir(train_high_folder))
print("Validation Low Images:", os.listdir(val_low_folder))
print("Validation High Images:", os.listdir(val_high_folder))
```

```python
from tensorflow.keras import layers, Model

def build_generator(input_shape=(256, 256, 3)):
    inputs = layers.Input(shape=input_shape)

    # Encoder
    e1 = layers.Conv2D(64, (3, 3), strides=2, padding="same", activation="relu")(inputs)
    e2 = layers.Conv2D(128, (3, 3), strides=2, padding="same", activation="relu")(e1)
    e3 = layers.Conv2D(256, (3, 3), strides=2, padding="same", activation="relu")(e2)

    # Bottleneck
    b = layers.Conv2D(512, (3, 3), padding="same", activation="relu")(e3)

    # Decoder
    d1 = layers.Conv2DTranspose(256, (3, 3), strides=2, padding="same", activation="relu")(b)
    d2 = layers.Conv2DTranspose(128, (3, 3), strides=2, padding="same", activation="relu")(d1)
    d3 = layers.Conv2DTranspose(64, (3, 3), strides=2, padding="same", activation="relu")(d2)

    outputs = layers.Conv2D(3, (3, 3), padding="same", activation="sigmoid")(d3)

    return Model(inputs, outputs)

generator = build_generator()
generator.summary()
```

```python
def build_discriminator(input_shape=(256, 256, 3)):
    inputs = layers.Input(shape=input_shape)

    d1 = layers.Conv2D(64, (3, 3), strides=2, padding="same", activation="relu")(inputs)
    d2 = layers.Conv2D(128, (3, 3), strides=2, padding="same", activation="relu")(d1)
    d3 = layers.Conv2D(256, (3, 3), strides=2, padding="same", activation="relu")(d2)
    d4 = layers.Conv2D(512, (3, 3), strides=2, padding="same", activation="relu")(d3)

    outputs = layers.Conv2D(1, (3, 3), padding="same", activation="sigmoid")(d4)

    return Model(inputs, outputs)

discriminator = build_discriminator()
discriminator.summary()
```

```python
from tensorflow.keras.optimizers import Adam

# Compile Discriminator
discriminator.compile(optimizer=Adam(0.0002, 0.5), loss="binary_crossentropy")

# GAN Model
def build_gan(generator, discriminator):
    discriminator.trainable = False
    gan_input = layers.Input(shape=(256, 256, 3))
    generated_image = generator(gan_input)
    gan_output = discriminator(generated_image)
    return Model(gan_input, [generated_image, gan_output])

gan = build_gan(generator, discriminator)
gan.compile(optimizer=Adam(0.0002, 0.5), loss=["mse", "binary_crossentropy"])
gan.summary()
```
